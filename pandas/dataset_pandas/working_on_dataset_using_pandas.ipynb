```python
# what is the avg tip given on each day?
# use groupby()
total_tip_avg = data.groupby("day")["tip"].mean()
```

```python
total_tip_avg
```

```python
# what is the total tip collected by each sex group?
# groupby
total_tip = data.groupby("sex")["tip"].sum()
total_tip
```

```python
# find the average total_bill for lunch and dinner separately?
avg_total_bill = data.groupby("time")["total_bill"].mean()
```

```python
avg_total_bill
```

```python
avg_total_bill["Lunch"]
```

```python
avg_total_bill["Dinner"]
```

```python
# which day had the highest total bill collected overall ? 
highest_bill_day = data.groupby("day")["total_bill"].sum().nlargest(1)
```

```python
highest_bill_day
```

```python
# group the data by smoker and calcuate the avg tip for each group 
grp_data_smoker = data.groupby("smoker")["tip"].mean()
```
